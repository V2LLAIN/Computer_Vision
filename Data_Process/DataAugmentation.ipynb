```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import tensorflow.keras.datasets as ds
```

```python
(X_train, y_train), (X_test, y_test) = ds.cifar10.load_data()
```

```python
X_train = X_train.astype('float32'); X_train /= 255.0
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## 앞의 15개의 이미지만 augmentation 적용

```python
X_train = X_train[0:15,]; y_train = y_train[0:15,]

class_name = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

#### trains set image출력

```python
plt.figure(figsize=(20,2))
plt.suptitle("First 15 images in the train set")

for i in range(15):
    plt.subplot(1, 15, i+1)
    plt.imshow(X_train[i])
    plt.xticks([]); plt.yticks([])
    plt.title(class_name[int(y_train[i])])
plt.show()
```

```python
batch_siz = 4    # 한번에 생성하는 양(mini-batch)

generator = ImageDataGenerator(rotation_range=20.0,
                              width_shift_range=0.2,
                               height_shift_range=0.2,
                               horizontal_flip=True
                              )

gen = generator.flow(X_train, y_train, batch_size=batch_siz)
```

```python
for i in range(3):
    img, label = gen.next()
    plt.figure(figsize=(8, 2.4))
    plt.suptitle("Generator Image" + str(i+1))
    
    for j in range(batch_siz):
        plt.subplot(1, batch_siz, j+1)
        plt.imshow(img[j])
        plt.xticks([]); plt.yticks([])
        plt.title(class_name[int(label[j])])
        
    plt.show()
```
